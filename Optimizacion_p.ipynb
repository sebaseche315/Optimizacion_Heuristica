### **Creación Función Qings**

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go
```

```
def qingfcn(x1min,x1max,x2min,x2max,R):
  x1=np.linspace(start=x1min,stop=x1max,num=R)
  x2=np.linspace(start=x2min,stop=x2max,num=R)
  #para 2 dimensiones
  f1=pd.DataFrame(((x1**2)-1)**2)
  fn=pd.DataFrame(((x1**2)-1)**2+((x2**2)-2)**2)
  x1=pd.DataFrame(x1)
  value_function=pd.concat([f1,fn,x1],axis=1)
  value_function.columns=['f1','fn','x1']
  return value_function
```

```
funcion_qingfcn=qingfcn(-2,2,-2,2,1500)
```

```
plt.plot(funcion_qingfcn['x1'], funcion_qingfcn['f1'])
plt.xlabel('X')
plt.ylabel('f(x)')
plt.title('2D View')
plt.show()
```

```
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(funcion_qingfcn['x1'], funcion_qingfcn['f1'],funcion_qingfcn['fn'],c='r', marker='o')
ax.set_xlabel('x1')
ax.set_ylabel('f1')
ax.set_zlabel('fn')

plt.show()
```

```
fig = go.Figure(data=[go.Surface(z=funcion_qingfcn.values)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_z=True))
fig.update_layout(title='Function Qings', autosize=False,
                  scene_camera_eye=dict(x=1.87, y=0.88, z=-0.64),
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90)
)

fig.show()
```
